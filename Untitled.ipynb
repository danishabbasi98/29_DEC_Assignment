```python
def abc(name,*number):#multiple Arguments passes and keyword always passes as a last Argument or first .not in middle and multiples
    print(name)
    print(number)
    print(sum(number))
abc('Qasim',5,4,7,8)
```

```python
def abc(*number,name):#multiple Arguments passes and keyword always passes as a last Argument
    print(name)
    print(number)
    abc(5,4,5,'Qasim')
```

```python
def abc(name,**kv):
    print(name)
    print(kv)
abc('Qasim',a='aa',b='bbbb',c='ccc')
```

Recursive_Function
---

```python
def abc(n):
    if n==1:
        return 1
    return n + abc(n-1)
abc(5)
#1*2*3*4*5
```

Generator Function
---

```python
def my_range(n):
    for i in range(1 , n+1):
        print(i)
my_range(10)
```

```python
def my_range(n):
    for i in range(1 , n+1):
        yield(i)
a = my_range(10)
print(next(a))
print(next(a))
print(next(a))
print(next(a))
```

```python
def abc(n):
    for i in range(1 , n+1):
        if(i%5 == 0 and i%7 == 0 and i%15 == 0 ):
            yield(i)#stop the loop.its executed one by one
a = abc(1000)
print(next(a))
print(next(a))
print(next(a))
list(a)
```
